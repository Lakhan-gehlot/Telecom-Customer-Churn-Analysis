#use function for convert citizen "0" to "no" and "1" to "yes"

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

        
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
plt.figure(figsize = (4,3))


ax = sns.countplot(x = 'Churn' , data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of customer by Churn")
plt.show()
```

```python
plt.figure(figsize = (2,3))


gb = df.groupby("Churn").agg({'Churn' : "count"})

plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")

plt.title("Percentage of Churn Customer" , fontsize = 10)

plt.show()
```

```python
plt.figure(figsize = (4,4)) 
ax = sns.countplot(x= "SeniorCitizen" , data = df,)
ax.bar_label(ax.containers[0])
plt.title("Count Of Customer By SeniorCitizen")
plt.show()
```

```python
# Step 1: Calculate counts and percentages
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
total_per_group = grouped.groupby('SeniorCitizen')['Count'].transform('sum')
grouped['Percent'] = grouped['Count'] / total_per_group * 100

# Step 2: Pivot for stacked bar chart
pivot_df = grouped.pivot(index='SeniorCitizen', columns='Churn', values='Percent').fillna(0)

# Step 3: Plot
plt.figure(figsize=(3, 4))
bars = plt.bar(pivot_df.index, pivot_df.iloc[:, 0], label=pivot_df.columns[0])
if pivot_df.shape[1] > 1:
    bars2 = plt.bar(pivot_df.index, pivot_df.iloc[:, 1], bottom=pivot_df.iloc[:, 0], label=pivot_df.columns[1])

# Step 4: Add percentage labels
for i in pivot_df.index:
    y_offset = 0
    for churn_status in pivot_df.columns:
        percent = pivot_df.loc[i, churn_status]
        if percent > 0:
            plt.text(i, y_offset + percent / 2, f'{percent:.1f}%', ha='center', va='center', fontsize=8)
            y_offset += percent

# Final touches
plt.xticks(pivot_df.index, ['Not Senior', 'Senior'])
plt.title("Churn by SeniorCitizen (% of Total)")
plt.ylabel("Percentage")
plt.legend(title="Churn", bbox_to_anchor = (0.9,0.9))
plt.tight_layout()
plt.show()
```

```python


# List of categorical columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid
rows = 3
cols_per_row = 3
fig, axes = plt.subplots(rows, cols_per_row, figsize=(15, 10))
axes = axes.flatten()

# Loop through each column and plot
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].tick_params(axis='x', rotation=45)

# Remove any unused subplots (if any)
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#The count plots display customer adoption across nine telecom service categories, such as Internet Service, Streaming TV, and Tech Support.
 Many services show a significant portion of customers opting out, especially where "No internet service" or "No phone service" is an option.
 Internet-based services like Online Security and Streaming TV exhibit varied uptake, suggesting potential areas for upselling or retention strategies
